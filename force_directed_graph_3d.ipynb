```python
import jgraph

graph = {
    'nodes': {},
    'edges': []
}

with open("./pdf2text/output/topics.txt", "r") as lines:
    i = 0
    limit = 50
    for line in lines:
        if i is limit:
            break
        i += 1
        linenum, topicnum, topics = line.split(',')

        graph['nodes'][linenum] = {'color': 0xff22ff}

        for topic in topics.split(' '):
            topic = topic.strip()
            graph['nodes'][topic] = {'color': 0x2222ff}
            edge = {
                'source': linenum,
                'target': topic
            }
            graph['edges'].append(edge)
```

```python
import json
json.dumps(graph)
```
